```python
from itertools import combinations
from bisect import bisect_right

def solve_bouquets(n, budgets):
    # Генерация всех степеней двойки до 2^100
    powers_of_two = [2**i for i in range(101) if 2**i <= 10**18]
    
    # Предобработка всех возможных сумм трёх различных степеней двойки
    possible_sums = []
    for comb in combinations(powers_of_two, 3):
        possible_sums.append(sum(comb))
    possible_sums.sort()
    
    results = []
    for budget in budgets:
        # Находим максимальную сумму <= budget
        idx = bisect_right(possible_sums, budget) - 1
        if idx >= 0:
            results.append(possible_sums[idx])
        else:
            results.append(-1)
            
    return results


n = int(input())
budgets = [int(input()) for _ in range(n)]

results = solve_bouquets(n, budgets)
print("\n".join(map(str, results)))
```

```python
def min_operations(n, x, y, z, a):
    # Упрощаем числа на основе кратности
    divisors = sorted([x, y, z], reverse=True)  # Сортируем по убыванию
    x, y, z = divisors[0], divisors[1], divisors[2]  # x >= y >= z

    # Убираем лишние числа, если одно кратно другому
    if x % y == 0:
        y = None  # y не имеет значения, так как x покрывает y
    if x % z == 0 or (y is not None and y % z == 0):
        z = None  # z не имеет значения, так как x или y покрывают z

    # Функция для подсчёта операций, чтобы сделать элемент делящимся на число
    def operations_to_make_divisible(val, divisor):
        if divisor is None:  # Игнорируем, если число уже покрыто
            return 0
        if val % divisor == 0:
            return 0
        return divisor - (val % divisor)

    # Минимальные значения для оставшихся чисел
    min_x = float('inf') if x is not None else 0
    min_y = float('inf') if y is not None else 0
    min_z = float('inf') if z is not None else 0

    # Рассчитываем минимальные операции для каждого элемента массива
    for ai in a:
        if x is not None:
            min_x = min(min_x, operations_to_make_divisible(ai, x))
        if y is not None:
            min_y = min(min_y, operations_to_make_divisible(ai, y))
        if z is not None:
            min_z = min(min_z, operations_to_make_divisible(ai, z))

    # Суммируем минимальные операции
    result = min_x + min_y + min_z
    
    return result


n, x, y, z = map(int, input().split())
a = list(map(int, input().split()))

print(min_operations(n, x, y, z, a))
```
