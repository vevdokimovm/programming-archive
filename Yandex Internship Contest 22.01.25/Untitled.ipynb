```python
def count_subarrays_with_median(n, b, arr):
    """
    Count the number of subarrays of odd length where the median equals b.

    :param n: Length of the array
    :param b: Target median value
    :param arr: The permutation array
    :return: Count of subarrays with the desired property
    """
    # Find the position of b in the array
    b_index = arr.index(b)

    # Balances for prefix sums
    prefix_balance = {0: 1}
    balance = 0
    count = 0

    # Traverse the array to compute balances
    for i in range(n):
        if arr[i] < b:
            balance -= 1
        elif arr[i] > b:
            balance += 1

        # When the current element is the median b
        if i >= b_index:
            count += prefix_balance.get(balance, 0)
            count += prefix_balance.get(balance - 1, 0)

        # Update prefix balances
        if i < b_index:
            prefix_balance[balance] = prefix_balance.get(balance, 0) + 1

    return count

# Input reading and processing
def main():
    N, B = [int(x) for x in input().split()]

    A = [int(x) for x in input().split()]

    result = count_subarrays_with_median(N, B, A)
    print(result)

if __name__ == "__main__":
    main()

# Test cases
def test():
    # Test case 1
    n1, b1, arr1 = 5, 2, [5, 4, 3, 2, 1]
    assert count_subarrays_with_median(n1, b1, arr1) == 2

    # Test case 2
    n2, b2, arr2 = 6, 3, [3, 6, 5, 4, 2, 1]
    assert count_subarrays_with_median(n2, b2, arr2) == 1

    # Test case 3
    n3, b3, arr3 = 7, 4, [5, 7, 2, 4, 3, 1, 6]
    assert count_subarrays_with_median(n3, b3, arr3) == 4

    print("All test cases passed!")

test()
```

```python
def minimize_years(N, A, B, C):
    years_to_remove = set(range(N)) # Все года, которые можно потенциально удалить
    current_A, current_B, current_C = set(A), set(B), set(C)
    final_intersection = current_A & current_B & current_C  # Начальное пересечение

    # Пока пересечение изменяется, повторяем процесс
    while True:
        new_years_to_remove = set()
        for i in years_to_remove:
            if A[i] not in final_intersection or B[i] not in final_intersection or C[i] not in final_intersection:
                new_years_to_remove.add(i)
        
        # Если нечего больше удалять, выходим
        if not new_years_to_remove:
            break

        # Убираем соответствующие годы и пересчитываем пересечение
        years_to_remove -= new_years_to_remove
        current_A = {A[i] for i in years_to_remove}
        current_B = {B[i] for i in years_to_remove}
        current_C = {C[i] for i in years_to_remove}
        final_intersection = current_A & current_B & current_C

    # Итоговое количество удалённых лет
    return N - len(years_to_remove)


N = int(input())
A = list(map(int, input().split()))
B = list(map(int, input().split()))
C = list(map(int, input().split()))

ans = minimize_years(N, A, B, C)
print(ans)
```

```python
def check_actions(query, vocab):
    actions = 0
    for word in vocab:
        actions += 1
        if word == query:
            actions += len(word)
            return actions
        
        s = len(word) if len(word) < len(query) else len(query)
        for i in range(s):
            if word[i] == query[i]:
                actions += 1
            else:
                break
    
    return actions


n = int(input())
vocab = []

for i in range(n):
    word = input()
    vocab.append(word)
    
q = int(input())
for i in range(q):
    word = input()
    actions = check_actions(word, vocab)
    print(actions)
```

```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end_of_word = False

class Trie:
    def __init__(self):
        self.root = TrieNode()

    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end_of_word = True

    def search_operations(self, query):
        node = self.root
        actions = 0
        prefix_length = 0

        for char in query:
            actions += 1  # Сравнение символа
            if char in node.children:
                node = node.children[char]
                prefix_length += 1
            else:
                break

            # Если нашли полное совпадение, добавляем длину префикса
            if node.is_end_of_word:
                actions += prefix_length
                return actions

        # Если префикс полностью обработан, добавляем его длину
        actions += prefix_length
        return actions + len(query) - prefix_length  # Учитываем оставшуюся часть запроса


def process_queries_with_trie(database, queries):
    trie = Trie()

    # Построение Trie для слов базы данных
    for word in database:
        trie.insert(word)

    # Обработка запросов
    results = []
    for query in queries:
        results.append(trie.search_operations(query))
    
    return results


N = int(input())
database = [input().strip() for _ in range(N)]
Q = int(input())
queries = [input().strip() for _ in range(Q)]

answer = process_queries_with_trie(database, queries)
print('\n'.join(map(str, answer)))
```
