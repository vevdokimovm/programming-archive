```python
r, c = [int(x) for x in input().split()]

crossword = [[0 for x in range(r)] for y in range(c)]

for i in range(r):
    row = list(input())
    print(row)
    for j in range(c):
        crossword[i][j] = row[j]
        
        
ans = []

# Сначала найдем все горизонтальные слова *->*
for i in range(r):
    # Зафиксировали строку и проходим по ней и находим все слова слева-направо
    # до разделителя '#' или конца строки     
    word = ''
    for j in range(c):
        if crossword[i][j] == '#':
            if len(word) >= 2:
                ans.append(word)
            word = ''
        
        if crossword[i][j] != '#':
            word += crossword[i][j]
    
    if len(word) >= 2:
        ans.append(word)
        

# Теперь пройдем также только сверху-вниз
for j in range(c):
    word = ''
    for i in range(r):
        if crossword[i][j] == '#':
            if len(word) >= 2:
                ans.append(word)
            word = ''
        
        if crossword[i][j] != '#':
            word += crossword[i][j]
        
    if len(word) >= 2:
        ans.append(word)
        
ans = sorted(ans)
# print(ans)
print(ans[0])

# for i in range(r):
#     for j in range(c):
#         print(crossword[i][j], end="")
#     print()
```
